# Read .csv from Multiple URLs with Pandas
###### https://www.football-data.co.uk/mmz4281/2122/E0.csv
###### https://www.football-data.co.uk/mmz4281/2122/E1.csv

##### Link: root + season + league

```python
#Above we see both links are same text except E0 and E1 is different so we can break the link

# link structure
'https://www.football-data.co.uk/mmz4281/' + '2122' + '/' + 'E0' + '.csv'       #breaking first link
```

```python
# create a root variable: Root variable is common part of links
root = 'https://www.football-data.co.uk/mmz4281/'
```

# Multiple leagues
###### . https://www.football-data.co.uk/mmz4281/2122/E0.csv                              
###### . https://www.football-data.co.uk/mmz4281/2122/E1.csv
###### . https://www.football-data.co.uk/mmz4281/2122/E2.csv
###### . https://www.football-data.co.uk/mmz4281/2122/E3.csv
###### . https://www.football-data.co.uk/mmz4281/2122/EC.csv

```python
# creating list of leagues
leagues = ['E0', 'E1', 'E2', 'E3', 'EC']
frame = []       #empty list to store csv files
# looping through leagues, read multiple csv and append it into a list
for league in leagues:
    df = pd.read_csv(root + '2122' + '/' + league + '.csv', encoding = 'unicode_escape')
    frame.append(df)
```

```python
# length of frames
len(frame)
```

```python
# show 1st, 2nd and 3rd element
frame[0]       #shows file of E0 league 2122 season
```

# Multiple Seasons

multiple season have change in season and league no. both

```python
for season in range(10,16):
    seasons = str(season)+str(season+1)
```

```python
#creating list of leagues
leagues = ['E0','E1','E2','E3','EC']
frames = []
# loop through range of seasons and insert new column
for league in leagues:
    for season in range(10,16):
        df = pd.read_csv(root + seasons + '/' + league + '.csv', encoding = 'unicode_escape')
        df.insert(1, 'season', season)      #season column at index 1
        frames.append(df)
```

```python
# total frames (6 seasons x 3 leagues)
len(frames)
```

```python
# show 1st, and last element (new season column was created)
frames[8]
```

# Organize All The Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {'Spanish La Liga':'SP1', 
                  'Spanish Segunda Division':'SP2',
                  'German Bundesliga':'D1',
                  'English Premier League':'E0', 
                  'English League 1':'E2', 
                  'English League 2':'E3' }
```

```python
# get a dictionary element
dict_countries['German Bundesliga']      #this gives value of this key 'German Bundesliga'
```

```python
dict_historical_data ={}                        #Initialize an empty dictionary to store historical data for each league
# looping through key elements

for league in dict_countries:
    frames = []       #one dataframe per league
    for season in range(10, 16):
        df = pd.read_csv(root + seasons + "/" + dict_countries[league]+ ".csv")
        df.insert(1, 'season', season)
        frames.append(df)
    df_frames = pd.concat(frames) # storing all seasons of a different league in a single dataframe using concat and store it in df_frame
    dict_historical_data[league] = df_frames # assigning dataframe to a new dictionary
```

```python
# show dataframe inside dict_historical_data
dict_historical_data['German Bundesliga']     #put any league name (like German Bundesliga) of which you want records of all season between  10 to 16
```
